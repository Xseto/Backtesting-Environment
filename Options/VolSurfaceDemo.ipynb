```python
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
from VolSurface import VolSurface
```

For data, this notebook uses call option chains on Citi from 1/31/2024

```python
def plottable_3d_info(label_sparsity_strike, label_sparsity_tenor, df: pd.DataFrame):
    """
    Transform Pandas data into a format that's compatible with plot_surface.
    """
    index = df.index
    columns = df.columns

    x, y = np.meshgrid(np.arange(len(columns)), np.arange(len(index)))
    z = np.array([[df[c][i] for c in columns] for i in index])
    
    xticks = dict(ticks=np.arange(len(columns))[::label_sparsity_tenor], labels=columns[::label_sparsity_tenor])
    yticks = dict(ticks=np.arange(len(index))[::label_sparsity_strike], labels=index[::label_sparsity_strike])

    return x, y, z, xticks, yticks
```

```python
file_path = 'Your own file path'
option_chains = pd.read_excel(file_path, sheet_name=None)
vol_surface = VolSurface(56.74, .01, np.log(1+.0368), datetime.strptime('01_31_2024', '%m_%d_%Y'), option_chains, 'E', 'C')
```

```python
vol_surface.implied_vol_surface()
```

```python
vol_surface.iv_surface_filled.head()
```

```python
# Plot implied vol surface
DF = vol_surface.iv_surface_filled

x, y, z, xticks, yticks = plottable_3d_info(5,4,DF)

# plot
fig = plt.figure()
axes = fig.add_subplot(projection='3d')
axes.plot_surface(x, y, z)

#axes.set_xlabel('Tenors')
axes.set_ylabel('Strikes')
axes.set_zlabel('Local Vol')
axes.set_zlim3d(bottom=0)
plt.xticks(**xticks)
plt.yticks(**yticks)

plt.show()
```
